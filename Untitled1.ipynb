```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as mp
```

```python
gpa_data=pd.read_csv("gpa_gender.csv")

x=gpa_data["Term"]
del gpa_data["Term"]


a=gpa_data["Female"]
b=gpa_data["Male"]
c=gpa_data["Entire Population"]


gpa_data
```

```python
#graph
mp.plot(a, linestyle="-.", linewidth=2, color="b")
mp.plot(b, linestyle=":", linewidth=2, color="g")
mp.plot(c, linewidth=2, color="r")

mp.xticks(np.arange(48), x, rotation=90, fontsize=7)
mp.title("Mean Undergraduate Cumulative GPA by Gender")
mp.ylabel("Avg GPA")
mp.legend(loc="lower right")
mp.show()
```
